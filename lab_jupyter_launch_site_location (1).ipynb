```python
from math import sin, cos, sqrt, atan2, radians

def calculate_distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0

    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)

    dlon = lon2 - lon1
    dlat = lat2 - lat1

    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))

    distance = R * c
    return distance
```

```python
# Funo para calcular distncia Haversine
from math import radians, sin, cos, sqrt, atan2
def calculate_distance(lat1, lon1, lat2, lon2):
    R = 6371.0  # raio da Terra em km
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat/2)**2 + cos(lat1)*cos(lat2)*sin(dlon/2)**2
    c = 2*atan2(sqrt(a), sqrt(1-a))
    return R*c

# Dados dos launch sites
launch_sites_df = pd.DataFrame({
    'Launch Site': ['CCAFS LC-40', 'VAFB SLC-4E', 'KSC LC-39A', 'CCAFS SLC-40'],
    'Lat': [28.562302, 34.632834, 28.573255, 28.561857],
    'Long': [-80.577356, -120.610746, -80.646895, -80.577366]
})
# Coordenadas das proximidades para cada site
proximities = {
    'CCAFS LC-40': {
        'Railway': (28.563197, -80.570775),
        'Highway': (28.56342, -80.57636),
        'Coastline': (28.563, -80.568),
        'City': (28.611, -80.807)
    },
    'VAFB SLC-4E': {
        'Railway': (34.633, -120.615),
        'Highway': (34.631, -120.609),
        'Coastline': (34.630, -120.605),
        'City': (34.639, -120.457)
    },
    'KSC LC-39A': {
        'Railway': (28.574, -80.645),
        'Highway': (28.576, -80.642),
        'Coastline': (28.571, -80.630),
        'City': (28.611, -80.807)
    },
    'CCAFS SLC-40': {
        'Railway': (28.563197, -80.570775),
        'Highway': (28.56342, -80.57636),
        'Coastline': (28.563, -80.568),
        'City': (28.611, -80.807)
    }
}

# Inicializa o mapa no centro dos sites
map_center = [launch_sites_df['Lat'].mean(), launch_sites_df['Long'].mean()]
site_map = folium.Map(location=map_center, zoom_start=6)
```

```python
# Para cada launch site
for i, row in launch_sites_df.iterrows():
    site_name = row['Launch Site']
    launch_coord = (row['Lat'], row['Long'])
    
    # Adiciona marcador do launch site
    folium.Marker(
        location=launch_coord,
        popup=f"{site_name} (Launch Site)",
        icon=folium.Icon(color='blue', icon='rocket', prefix='fa')
    ).add_to(site_map)
    
    # Para cada proximidade no site
    for proximity_name, proximity_coord in proximities[site_name].items():
        # Calcula distncia
        dist = calculate_distance(launch_coord[0], launch_coord[1], proximity_coord[0], proximity_coord[1])
        
        # Adiciona marcador com a distncia
        folium.Marker(
            location=proximity_coord,
            icon=DivIcon(
                icon_size=(150, 20),
                icon_anchor=(0, 0),
                html=f'<div style="font-size: 12px; color:#d35400;"><b>{proximity_name}: {dist:.2f} KM</b></div>'
            )
        ).add_to(site_map)
        
        # Desenha linha entre launch site e proximidade
        folium.PolyLine(
            locations=[launch_coord, proximity_coord],
            weight=2,
            color='green',
            opacity=0.7
        ).add_to(site_map)

# Mostrar mapa
site_map
```
